```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# Functional method
```

```python
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, "r")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")

plt.subplot(1,2,2)
plt.plot(x,y,"b")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")

```

```python
# object oriented language
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y, "g")
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.set_xlabel("X label")
axes1.set_ylabel("Y label")
axes1.set_title("LARGER PLOT")
axes2.set_title("SMALLER PLOT")

axes1.plot(x,y, "r")
axes2.plot(y,x, "y")
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)
# axes.plot(x,y)

# for current_ax in axes:
#     current_ax.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title("FIRST PLOT")
axes[1].plot(y,x)
axes[1].set_title("SECOND PLOT")

plt.tight_layout()
```

```python
# FIGURE SIZE AND DPI
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title("Title")
ax.plot(x,x**2, label = "X squared")
ax.plot(x,x**3, label = "Y Cubed")

ax.legend(loc=(0.1,0.1))
```

```python
# Meta Part3
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = "purple", linewidth =3,alpha=0.5, ls = '--',marker = "o",markersize =20,
       markerfacecolor = "yellow", markeredgewidth = 3, markeredgecolor ="green") 
# alpha is used to transparent the line

```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color ="purple", lw = 2, ls = "--")
ax.set_xlim([0,1])
ax.set_ylim
```
